```python
## Make a plot from dictionary
butter_prices = {"Almond Butter" : 10,
                 "Peanut Butter" : 8,
                 "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar( butter_prices.keys(), butter_prices.values())
ax.set(title = "My Store", ylabel = "Price") ## title gives the title and y_label give the label to the y axis
```

## Which one to use? Pyplot vs Matplotlib OO method
1. When plotting something quickly, okay to use pyplot
2. When Plotting something more advanced, use the OO method


```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(heart_disease)
```

```python
over_50.plot(kind = "scatter",
             x = "age",
             y = "chol",
             c = "target");
```

## OO Method from sratch

```python
fig, ax = plt.subplots(figsize = (10, 6))

## Plot the data

scatter  = ax.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50['target'])

## customize the plot
ax.set(title="Heart Disease",
       xlabel = "Age",
       ylabel = "Cholestrol"  );


## Add a legend

ax.legend(*scatter.legend_elements(), title= "Target");
```

```python
over_50.target.values
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize= (10, 10))

# Add data to ax0
scatter  = ax0.scatter(x = over_50["age"],
                       y = over_50["chol"],
                       c = over_50["target"],
                       cmap = "winter")

## customize ax0
ax0.set(title = "Heart Disease and Cholesterol Levels",
        xlabel = "Age",
        ylabel = "Cholesterol")

# change the axis limits
ax0.set_xlim([50, 80])

## Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

## Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle = "--")

## Add data to ax1
scatter  = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                       cmap = "winter"
                       )

## customize ax1
ax1.set(title="Heart Disease",
        xlabel = "Age",
        ylabel = "Max Heart rate")

## Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title= "Target")

## Add a mealine
ax1.axhline(y = over_50["thalach"].mean(),
            linestyle= "--");
```
